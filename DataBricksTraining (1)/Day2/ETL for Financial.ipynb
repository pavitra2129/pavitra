```python
df_fin=spark.read.csv(input_path,header=True,inferSchema=True)
```

```python
for column in df_fin.columns:    
    df_fin.withColumnRenamed(column, column.lower())
```

```python
new_col=['transaction_id',  'customer_id',  'account_type',  'transaction_type', 'amount_usd','branch', 'transaction_date']
df1=df_fin.toDF(*new_col).withColumn("ingestion_date", current_timestamp()).withColumn("path",col("_metadata.file_name"))
```
