## Types of views

A virtual table 
1. standard view ( persisted save to schema)
2. temp view ()

```python
%sql
create view acc_dev.naval.top_customers as 
select customerid,sum(total_amount) as total_amount from acc_dev.naval.sales group by customerid sort by total_amount desc limit 3
```

```python
%sql
create temp view last_custom_view as 
select customerid,sum(total_amount) as total_amount from acc_dev.naval.sales group by customerid sort by total_amount limit 3
```

### UDF

```python
%sql
create function function_name(para datatype)
returns datatypes
return logic
```

```python
%sql
create function acc_dev.pavitra.full_name(first_name string, last_name string)
returns string
return concat(first_name, ' ', last_name)
```

```python
%sql
select acc_dev.pavitra.full_name("pavitra","battina") as name
```

```python
%sql
create function acc_dev.pavitra.sum_number(first_number int, second_number int)
returns int
return first_number + second_number
```

```python
%sql
select acc_dev.pavitra.sum_number(1234,2344) as sum
```
